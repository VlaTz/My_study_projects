### Вывод

Таблица состоит из 23699 строк, и 22 столбцов. В 14 из них обнаружены пропуски, с которыми еще предстоит разобраться. В большинстве случаев, пропуски не несут ничего критичного. Потому что пропуск, в столбце, например, balcony, может означать отсутствие балкона. Вызваны пропуски могут быть причиной не заполнения соответсствущих колонок как раз по причине отсутствия (нет балкона - зачем заполнять). В столбцах, где у данных булевое значение, пропуски могут означать False. Остальноые пропуски могут быть по причине не внимательного заполнения данных о квартире (человеческий фактор). Нужно изучать каждый столбец.

## Предобработка данных

###  Обработка пропусков

```python
df.isna().sum() #считаем, сколько есть пропусков и где они
```

Изучим каждый столбец и избавимся от пропусков в нем.

```python
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median()) #в столбце ceiling_height(высота потолков)
#заменим значения на медианные
```

Пропусков в столбце floors_total(кол-во этажей в доме) всего 86 шт. Это 0,36% от всех строк. Так как их меньше 5%, их можно удалить со спокойной душой. Ведь данные о кол-ве этажей не заменить на медиану или среднее. Это может повлиять на то будет ли квартира на первом или последнем этаже. Аналогично поступаем с locality_name(название населённого пункта). Там пропусков 49, это 0,21% от всех строк. Данные о локации тоже не заменить медианой или средним. Их удаление не исказит данные и не повредит ходу исследования.

```python
df = df.dropna(subset=['floors_total']) #удаляем прстроки с пропусками в floors_total и locality_name
df = df.dropna(subset=['locality_name']).reset_index(drop=True)
```

Для заполнения пропусков в living_area (жилая площадь) группируем квартиры по количеству комнат, и считаем для них среднюю площадь. После подсчета заполним пропуски средним значениям. У пропущенных значений будем смотреть, скольки комнатная квартира и заполнять соответсвенно - если пропуск у однокомнотной квартиры, заполним средним значением для однокомнатных квартир.

Для этого сделаем сводную таблицу:

```python
df_rooms = df.pivot_table(index='rooms', values='living_area') #aggfunc по умолчанию применят mean()
display(df_rooms)
```

Здесь видны выбросы. Квартиры с большим количеством комнат, разберемся с ними позже. Где количество комнат 0, примем за квартиру-студию.

```python
df['living_area'] = df['living_area'].fillna(0) #заполняем пропуски в 'living_area' нулями
```

```python
for i in range(len(df)):
    if df.loc[i, 'living_area']==0:
        if df.loc[i, 'rooms']==0:
            df.loc[i, 'living_area']=df_rooms.loc[0, 'living_area']
        elif df.loc[i, 'rooms']==1:
            df.loc[i, 'living_area']=df_rooms.loc[1, 'living_area']
        elif df.loc[i, 'rooms']==2:
            df.loc[i, 'living_area']=df_rooms.loc[2, 'living_area']
        elif df.loc[i, 'rooms']==3:
            df.loc[i, 'living_area']=df_rooms.loc[3, 'living_area']
        elif df.loc[i, 'rooms']==4:
            df.loc[i, 'living_area']=df_rooms.loc[4, 'living_area']
        elif df.loc[i, 'rooms']==5:
            df.loc[i, 'living_area']=df_rooms.loc[5, 'living_area']
        elif df.loc[i, 'rooms']==6:
            df.loc[i, 'living_area']=df_rooms.loc[6, 'living_area']
        elif df.loc[i, 'rooms']==7:
            df.loc[i, 'living_area']=df_rooms.loc[7, 'living_area']
        elif df.loc[i, 'rooms']==8:
            df.loc[i, 'living_area']=df_rooms.loc[8, 'living_area']
        elif df.loc[i, 'rooms']==9:
            df.loc[i, 'living_area']=df_rooms.loc[9, 'living_area']
                
```

Выше я написал условие для замены пропусков в столбце living_area. Если значение в столбце living_area равно 0: проверяем, чеиу равно значение в rooms в этой же строчке. Если равно 1, заполняем средним значением для однакомнотных квартир, если 2, то для средним для двухкомнатных и тд. Всего условия написано для 5 комнатных квартир. Проверим наличие пропусков после их заполнения.

```python
cnd=0 #проверим сколько нулей в столбце 'living_area'- если их нет, то все пропуски заполнены
for i in df['living_area']:
    if i==0:
        cnd+=1
print(cnd)
```

```python
df['is_apartment'].value_counts() #проверяем, есть ли в 'is_apartment' заполненные ячейки
```

Логично утверждать, что пропущенные значения равны False. И означают, что квартира не апартаменты.

```python
df['is_apartment'] = df['is_apartment'].fillna(False) #заменим пропущенные значения на Flase
```

```python
df_kitchen = df.pivot_table(index='rooms', values='kitchen_area') #для kitchen_area используем тот 
display(df_kitchen) #же прием, что и для 'living_area', заполним пропуски по средниму значению от количества комнат.
```

```python
df['kitchen_area'] = df['kitchen_area'].fillna(0)
for i in range(len(df)):
    if df.loc[i, 'kitchen_area']==0:
        if df.loc[i, 'rooms']==1:
            df.loc[i, 'kitchen_area']=df_kitchen.loc[1, 'kitchen_area']
        elif df.loc[i, 'rooms']==2:
            df.loc[i, 'kitchen_area']=df_kitchen.loc[2, 'kitchen_area']
        elif df.loc[i, 'rooms']==3:
            df.loc[i, 'kitchen_area']=df_kitchen.loc[3, 'kitchen_area']
        elif df.loc[i, 'rooms']==4:
            df.loc[i, 'kitchen_area']=df_kitchen.loc[4, 'kitchen_area']
        elif df.loc[i, 'rooms']==5:
            df.loc[i, 'kitchen_area']=df_kitchen.loc[5, 'kitchen_area']
        elif df.loc[i, 'rooms']==6:
            df.loc[i, 'kitchen_area']=df_kitchen.loc[6, 'kitchen_area']
        elif df.loc[i, 'rooms']==7:
            df.loc[i, 'kitchen_area']=df_kitchen.loc[7, 'kitchen_area']
        elif df.loc[i, 'rooms']==8:
            df.loc[i, 'kitchen_area']=df_kitchen.loc[8, 'kitchen_area']
        elif df.loc[i, 'rooms']==9:
            df.loc[i, 'kitchen_area']=df_kitchen.loc[9, 'kitchen_area']
                
```

```python
cnd=0 #проверим сколько нулей в столбце 'kitchen_area'
for i in df['kitchen_area']:
    if i==0:
        cnd+=1
print(cnd)
```

Заменим типы данных в floors_total и balcony. Т.к. числа в этих столбцах можно округлить. В стольбцах с пропущенными значениями формат поменять нельзя. В стобцах, таких как ceiling_height и floors_total и т.д., важна дробная часть. Поэтому оставим их.

## Расчёты и добавление результатов в таблицу

```python
df['meter_price'] = df['last_price']/df['total_area'] #добавление в таблицу столбца со стоимостью квадратного метра
```

```python
df['day_start'] = pd.DatetimeIndex(df['first_day_exposition']).day #добавление в таблицу столбца с днем стартра продажи
df['month_start'] = pd.DatetimeIndex(df['first_day_exposition']).month #с месяцем старта продажи
df['year_start'] = pd.DatetimeIndex(df['first_day_exposition']).year #с годом старта продажи
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
df['day_of_week'] = df['first_day_exposition'].dt.weekday #c днем недели
```

Для категоризации квартир по этажам напишем функцию, которая бдует работать по двум столбцам 'floor' и 'floors_total'

```python
row_values = [1, 10]
row_columns = ['floor', 'floors_total']
row = pd.Series(data=row_values, index=row_columns)
def floor(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor==1:
        return 'первый'
    if floor==floors_total:
        return 'последний'
    return 'другой'
df['floor_info']=df.apply(floor, axis=1)
```

```python
df['floor_info'].value_counts() #проверим, сработала ли функция
```

```python
df['living_to_all'] = df['living_area']/df['total_area'] #добавляем столбец с отношением жилой площади к общей
df['kitchen_to_total'] = df['kitchen_area']/df['total_area']#и столбец отношения кухонной площади к общей
```

### Какие факторы больше всего влияют на стоимость квартиры

#### Площадь, число комнат, этаж, удалённость от центра

Для ответа на этот вопрос нужно построить матриццу диаграм рассеяния. Для этого необходимо исключить лишние столбцы.

```python
pd.set_option('chained_assignment', None)
df_price_answer = df_new[['last_price', 'total_area', 'rooms', 'floor_info', 'cityCenters_nearest']]
def floor_index(floor): #заменим информацию об этаже на индексы
    if floor=='первый':
        return 0
    if floor=='последний':
        return 2
    return 1
df_price_answer.loc[:, 'floor_info']=df_price_answer.loc[:,'floor_info'].apply(floor_index)
```

```python
pd.plotting.scatter_matrix(df_price_answer, figsize=(20,  20))
```

Цена зависит от квадратного метра: чем больше площадь, тем выше цена

Цена зависит от количества комнат: цены растут, если растет кол-во комнат. После 5 комнат, цены падают, кол-во комнат увеличивается

Цена зависит от этажа: квартиры не на первом и не на последнем этаже дороже остальных. Квартиры на первом и последнем этаже стоят почти одинаково

Цена немного зависит от удаленности от центра: не все квартиры, которые ближе к центру дороже более удаленых квартир. Причем их меньшинство
